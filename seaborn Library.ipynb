```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('diamonds')
```

```python
df.head()
```

here all the feature except carat is independent feature and tip is dependent to other independent feature

# corelation with heatmap

corelation is the relation between two feature which are numeric value the corelation ranges from -1 to 1

```python
df.corr(min_periods=1)
```

```python
sns.heatmap(df.corr())
```

Rightside the color indicator is shown which represents the scale of corelation value

# Jointplot

```python
sns.jointplot(x='carat',y='depth',data=df,kind='scatter')
```

the graph in up and right represents the histogram of x  and y axis respectively

```python
sns.jointplot(x='carat',y='depth',data=df,kind='reg')
```

a regration line is drawn 

# Pair plot

In this all the variables are paired with other all variable

```python
df=sns.load_dataset('tips')
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dis plot

gives histogram the count of values

```python
sns.displot(df['tip'])
```

```python
sns.displot(df['tip'],kde=True,bins=10)
```
